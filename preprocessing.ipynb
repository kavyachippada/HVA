```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv')
df
```

```python
df.info()
```

```python
sns.heatmap(df.isnull())
plt.show()
```

```python
print('Released_Year:\n',df['Released_Year'].unique())
print('Certificate:\n',df['Certificate'].unique())
```

```python
df=df[df['Released_Year']!="PG"]
df=df[df['Certificate'].notna()]
```

```python
print('Released_Year:\n',df['Released_Year'].unique())
print('Certificate:\n',df['Certificate'].unique())
```

```python
most_frequent=df['Certificate'].mode()[0]
df['Certificate']=df['Certificate'].fillna(most_frequent)
```

```python
most_frequent=df['Meta_score'].mode()[0]
df['Meta_score']=df['Meta_score'].fillna(most_frequent)
```

```python
most_frequent=df['Gross'].mode()[0]
df['Gross']=df['Gross'].fillna(most_frequent)
```

```python
print(df['Certificate'].isnull().sum())
print(df['Meta_score'].isnull().sum())
print(df['Gross'].isnull().sum())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Runtime']=df['Runtime'].str.replace('min','')
df['Gross']=df['Gross'].str.replace(',','')
```

```python
df['Released_Year'] = df['Released_Year'].astype(int)
df['Meta_score'] = df['Meta_score'].astype(int)
df['Gross'] = df['Gross'].astype(int)
df['Runtime'] = df['Runtime'].astype(int)
```

```python
df.info()
```

```python
df.to_csv('cleaned_imdb_dataset.csv',index=False)
```
